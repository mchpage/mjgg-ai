```python
# race_results_2025_race_results.csv load
import pandas as pd

df = pd.read_csv('race_results_2025_race_results.csv')
```

```python
df[['hrName', 'hrNo']].value_counts()
```

```python
# ord에 따른 새로운 파생변수 생성
'''
새로운 변수: prize_money

작동방식: 각 경주마다 순위(ord)에 따라 상금(chaksun)을 추가
조건:
- ord가 1인 경우: prize_money = chaksun1
- ord가 2인 경우: prize_money = chaksun2
- ord가 3인 경우: prize_money = chaksun3
- ord가 4인 경우: prize_money = chaksun4
- ord가 5인 경우: prize_money = chaksun5
- ord가 6 이상인 경우: prize_money = 0

'''

# hrName
df['prize_money'] = df.apply(
    lambda row: row['chaksun1'] if row['ord'] == 1 else
                row['chaksun2'] if row['ord'] == 2 else
                row['chaksun3'] if row['ord'] == 3 else
                row['chaksun4'] if row['ord'] == 4 else
                row['chaksun5'] if row['ord'] == 5 else
                0,
    axis=1
)
```

```python
# 말마다(hrName, hrNo) 상금 합계 계산 후 크기 순 정렬
horse_prize_money_sum = df.groupby(['hrName', 'hrNo'])['prize_money'].sum().reset_index()
horse_prize_money_sum = horse_prize_money_sum.sort_values(by='prize_money', ascending=False).reset_index(drop=True)

# 기수마다 상금 합계 계산
jockey_prize_money_sum = df.groupby(['jkName', 'jkNo'])['prize_money'].sum().reset_index()
jockey_prize_money_sum = jockey_prize_money_sum.sort_values(by='prize_money', ascending=False).reset_index(drop=True)
```
